```python
Q1=np.percentile(pima_df['pres'], 25)
Q1
```

```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3= np.percentile(col,[25,75])
    IQR=Q3-Q1
    Lower_range=Q1-(1.5*IQR)
    Upper_range=Q3+ (1.5*IQR)
    return Lower_range, Upper_range
```

```python
for column in pima_df.columns:
    lr,ur=remove_outlier(pima_df[column])
    pima_df[column]=np.where(pima_df[column]>ur,ur, pima_df[column])
    pima_df[column]=np.where(pima_df[column]<lr,lr,pima_df[column])
```

```python
pima_df.boxplot()
plt.figure(figsize=(10,10))
```

```python
x=pima_df.drop('class',axis=1)
x
```
